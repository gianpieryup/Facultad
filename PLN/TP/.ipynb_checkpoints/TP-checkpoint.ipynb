```python
import os
lista_archivos = os.listdir('./dataset-nlp-plagio-utn/')
path_archivos = ['dataset-nlp-plagio-utn/'+ a for a in lista_archivos]
path_archivos
```

```python
# Para leer la informacion de archivos variados (PDF, Word, pptx)
# use 'textract' porque usando 'open' tenia muchos problemas con el formato algunos decia que no eran 'utf-8'

# conda install -c conda-forge textract

# hay problemas con los archivos que tienen espacios en blanco en su nombre
# conda install -c conda-forge pdftotext 

# https://github.com/deanmalmgren/textract/issues/194
# Tambien estube sufriendo con el error del 'antiword' hay que descargarlo ponerlo en C:\ y agregarlo como variable de entorno
import textract

# https://www.nltk.org/api/nltk.tokenize.html
from nltk import tokenize

import re
def preparar_datos(path):
    # Extrae toda la data del archivo
    text = textract.process(path).decode("utf-8")
    
    # Considere el salto de linea como fin de la oracion, 
    # Reemplazo "\r\n" | "\n" por el ('. ') ya que es uno de los tokenizador usa esa combinacion como separador
    # \x0c and \xa0 tambien los limpio
    texto_sin_punto = text.replace(".","")
    texto_limpio = re.sub("(\n)+(\xa0)*|(\r\n)+(\x0c)*|(\u200b)|(\t)+", ". ", texto_sin_punto)
    
    # Tokenizamos por oraciones
    # Por palabra no tiene sentido
    texto_tokenizado = tokenize.sent_tokenize(texto_limpio)
    
    # Filtramos por frases mas largas(si considero las cortas nunca termina)
    texto_reducido = [oracion for oracion in  texto_tokenizado if len(oracion)>=20] 
    
    return texto_tokenizado
```

```python
# Usando la ejercitacion del TP3
# Donde Quitamos las stop_words y Lematizamos
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords

# ****************************************************************************
spanish_stemmer=nltk.stem.SnowballStemmer('spanish')
class StemmedCountVectorizer(CountVectorizer):
    def build_analyzer(self):
           analyzer=super(StemmedCountVectorizer,self).build_analyzer()
           return lambda doc: (spanish_stemmer.stem(w) for w in analyzer(doc))
    
stem_vectorizer = StemmedCountVectorizer(min_df=1, stop_words=stopwords.words('spanish'))
stem_analyze = stem_vectorizer.build_analyzer()
# *****************************************************************************
```

```python
def similitud_coseno(X_set,Y_set):
    l1 =[]
    l2 =[] 
    rvector = X_set.union(Y_set)
    
    for w in rvector: 
        if w in X_set: l1.append(1) 
        else: l1.append(0) 
        if w in Y_set: l2.append(1) 
        else: l2.append(0) 
    c = 0
    
    for i in range(len(rvector)): 
        c+= l1[i]*l2[i] 
    
    denominator = float((sum(l1)*sum(l2))**0.5)
    
    if (denominator == 0.0 ):
        return 0.0
    else :
        cosine = c / denominator
        return cosine
```

#### Obtener informacion de la caratula

Podemos usar este caracter para obtener la caratula de todo el texto

\x0c – the form feed character that forces a printer to move the next sheet of paper.

> texo.split('\x0c')[0]

```python

def validacion(oracion):
    letras_invalidas = ['@','[', 'ingeniería en sistemas de información', 'utn', 'trabajo', 'capitulo', 'fecha','legajo','cuatrimestre','catedra','ayudante','profesor','instructor', 'universidad', 'facultad', 'curso', 'email','integrante','alejandro prince','hernán borré','maximiliano bracho','cátedra']
    for l in letras_invalidas:
        if (l in oracion):
            return False
    # quitamos las oraciones muy cortas y muy largas  
    if (len(oracion) <3 or len(oracion) > 100):
        return False
    return True
```

```python
def obtener_caratula(path):
    caratula = textract.process(path).decode("utf-8")
    caratula_sin_blancos = re.sub("  +","",caratula)
    caratula_sin_punto = caratula_sin_blancos.replace(".","")
    caratula_limpio = re.sub("(\n)+(\xa0)*|(\r\n)+(\x0c)*|(\u200b)|(\t)+", ". ", caratula_sin_punto)
    return caratula_limpio.split('.')
```

```python
def obtener_datos_alumno(path):
    ''' La mayoria de los textos tienen un formato
        ***texto***.(palabra_clave).Nombre Alumno
        
        Donde palabra_clave in ['alumn','Datos del alumno' ,'nombre']
    '''
    caratula = obtener_caratula(path)

    #print(ct)
    #print('-'*15)
    filtrado = [o for o in caratula if validacion(o.lower())]
    print(filtrado)
    
    index_alumn = 0
    for i,oracion in enumerate(filtrado):
        if ('alumn' in oracion.lower()):
            if(len(oracion)>= 12):
                print(oracion)
            else:
                index_alumn= i
                
        if ('Datos del alumno' in oracion):
            index_alumn= i
        
        if ('nombre' in oracion.lower()):
            index_alumn= i
        
    if(index_alumn !=0):
        print('Alumno:',filtrado[index_alumn + 1])
        #El tema lo pongo como la union de todo lo demas hasta llegar a alumno
        # pero are un caso de if si es el primero para unir solo lo que resta
    
    print('Tema:',filtrado[0])
    print()
```

```python
def plagio(path, confianza):
  
    texto_analizar = preparar_datos(path)
    print('Nombre del archivo de texto procesado:',path[23:])
    
    obtener_datos_alumno(path)
    
    contador_oraciones_plagiadas = 0
    cantidad_total_de_oraciones = len(texto_analizar)
    
    for oracion_analizar in texto_analizar:
        print()
        print("-"*15)
        print("Oracion analizada:",oracion_analizar)
        print("-"*15)
        oracion_lematizada = set(stem_analyze(oracion_analizar))
        flag_copia=0
        
        for path_a in path_archivos:
            texto = preparar_datos(path_a)
            
            for index_oracion, oracion in enumerate(texto):
                oracion_lematizado_corpus = set(stem_analyze(oracion))
                
                if (similitud_coseno(oracion_lematizado_corpus,oracion_lematizada) >= confianza) :                    
                    print("Texto:",path_a[23:])
                    print("Oracion plagiada:",oracion ,'| ubicacion:', index_oracion)
                    flag_copia=1
                    
        if(flag_copia==1):
            contador_oraciones_plagiadas += 1
    
    print("Porcentaje de plagio del texto en general:", contador_oraciones_plagiadas/cantidad_total_de_oraciones)
```

```python
# Para que el resultado anterior de distinto de 1.0 quitar el mismo documento
plagio('dataset-nlp-plagio-utn/TP5 - Plataformas.docx',0.8)
```
